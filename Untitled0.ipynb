```python
import ccxt
import pandas as pd
import ta
import time
from datetime import datetime

exchange = ccxt.kraken()
coins = ['BTC/USD', 'ETH/USD', 'SOL/USD']

def check_signals():
    print(f"\n{'='*40}")
    print(f"Scan time: {datetime.now().strftime('%H:%M:%S')}")
    print(f"{'='*40}")

    for coin in coins:
        ohlcv = exchange.fetch_ohlcv(coin, '1h', limit=100)
        df = pd.DataFrame(ohlcv, columns=['time','open','high','low','close','volume'])
        df['rsi'] = ta.momentum.RSIIndicator(df['close'], window=14).rsi()

        price = df['close'].iloc[-1]
        rsi = round(df['rsi'].iloc[-1], 2)

        if rsi < 30:
            signal = "*** BUY ***"
        elif rsi > 70:
            signal = "*** SELL ***"
        else:
            signal = "HOLD"

        print(f"{coin}: ${price} | RSI: {rsi} | {signal}")

# Har 1 ghante me check karo (3600 seconds)
# Testing ke liye abhi 30 seconds rakha hai
print("Scanner shuru ho gaya! Ctrl+C dabao band karne ke liye.")

while True:
    check_signals()
    print("\nAgla scan 30 seconds baad...")
    time.sleep(30)
```

```python
import ccxt
import pandas as pd
import ta

exchange = ccxt.kraken()

# 500 ghante ka historical data
ohlcv = exchange.fetch_ohlcv('BTC/USD', '1h', limit=8760)
df = pd.DataFrame(ohlcv, columns=['time','open','high','low','close','volume'])

# RSI calculate karo
df['rsi'] = ta.momentum.RSIIndicator(df['close'], window=14).rsi()

# Strategy logic
balance = 10000  # $10,000 se shuru
btc = 0          # Shuru me 0 BTC
trades = []

for i in range(1, len(df)):
    price = df['close'].iloc[i]
    rsi = df['rsi'].iloc[i]

    # BUY signal — RSI 30 se kam, aur hamare paas BTC nahi
    if rsi < 35 and btc == 0 and balance > 0:
        btc = balance / price
        balance = 0
        trades.append({'type': 'BUY', 'price': round(price, 2), 'rsi': round(rsi, 2)})

    # SELL signal — RSI 70 se zyada, aur hamare paas BTC hai
    elif rsi > 65 and btc > 0:
        balance = btc * price
        btc = 0
        trades.append({'type': 'SELL', 'price': round(price, 2), 'rsi': round(rsi, 2)})

# Final value calculate karo
final_price = df['close'].iloc[-1]
if btc > 0:
    final_value = btc * final_price
else:
    final_value = balance

profit = round(final_value - 10000, 2)
profit_pct = round((profit / 10000) * 100, 2)

print("=" * 45)
print("BACKTEST RESULTS — BTC/USD (500 hours)")
print("=" * 45)
print(f"Starting balance : $10,000")
print(f"Final value      : ${round(final_value, 2)}")
print(f"Total profit     : ${profit}")
print(f"Return           : {profit_pct}%")
print(f"Total trades     : {len(trades)}")
print("=" * 45)
print("\nTrade history:")
for t in trades:
    print(f"  {t['type']} at ${t['price']} | RSI: {t['rsi']}")
```

```python
import ccxt
import pandas as pd
import ta
import time

exchange = ccxt.kraken()

# Data chunks me fetch karo
print("Data fetch ho raha hai... thoda wait karo")

all_data = []
timeframe = '1h'
symbol = 'BTC/USD'
limit = 500

# Pehla batch
ohlcv = exchange.fetch_ohlcv(symbol, timeframe, limit=limit)
all_data.extend(ohlcv)

# Baaki batches
for i in range(16):  # 16 x 500 = 8000+ candles ~ 1 saal
    since = all_data[0][0] - (limit * 3600 * 1000)  # Pehle candle se peeche jao
    ohlcv = exchange.fetch_ohlcv(symbol, timeframe, since=since, limit=limit)
    if not ohlcv:
        break
    all_data = ohlcv + all_data
    time.sleep(1)  # Server pe load na pade
    print(f"Batch {i+1}/16 fetch hua...")

df = pd.DataFrame(all_data, columns=['time','open','high','low','close','volume'])
df = df.drop_duplicates('time').sort_values('time').reset_index(drop=True)

print(f"\nTotal candles: {len(df)}")
print(f"Data from: {pd.to_datetime(df['time'].iloc[0], unit='ms').strftime('%d %b %Y')}")
print(f"Data to:   {pd.to_datetime(df['time'].iloc[-1], unit='ms').strftime('%d %b %Y')}")
```

```python
import yfinance as yf
import pandas as pd
import ta

# 2 saal ka data
btc = yf.download('BTC-USD', start='2023-01-01', end='2024-12-31', interval='1d', auto_adjust=True)
btc.columns = btc.columns.droplevel(1)

# RSI calculate karo
btc['rsi'] = ta.momentum.RSIIndicator(btc['Close'], window=14).rsi()

# Backtest
balance = 10000
btc_held = 0
trades = []

for i in range(1, len(btc)):
    price = btc['Close'].iloc[i]
    rsi = btc['rsi'].iloc[i]

    if rsi < 35 and btc_held == 0 and balance > 0:
        btc_held = balance / price
        balance = 0
        trades.append({'type': 'BUY', 'price': round(float(price), 2), 'rsi': round(float(rsi), 2)})

    elif rsi > 65 and btc_held > 0:
        balance = btc_held * price
        btc_held = 0
        trades.append({'type': 'SELL', 'price': round(float(price), 2), 'rsi': round(float(rsi), 2)})

# Final value
final_price = float(btc['Close'].iloc[-1])
final_value = btc_held * final_price if btc_held > 0 else balance
profit = round(final_value - 10000, 2)
profit_pct = round((profit / 10000) * 100, 2)

# Results
print("=" * 45)
print("BACKTEST — BTC/USD | 2 Saal | Daily")
print("=" * 45)
print(f"Starting balance : $10,000")
print(f"Final value      : ${round(final_value, 2)}")
print(f"Total profit     : ${profit}")
print(f"Return           : {profit_pct}%")
print(f"Total trades     : {len(trades)}")
wins = sum(1 for i in range(1, len(trades), 2) if i < len(trades) and trades[i]['price'] > trades[i-1]['price'])
print(f"Winning trades   : {wins}/{len(trades)//2}")
print("=" * 45)
print("\nTrade history:")
for t in trades:
    print(f"  {t['type']} at ${t['price']} | RSI: {t['rsi']}")
```

```python
import ccxt
import pandas as pd
import ta
import time
from datetime import datetime

exchange = ccxt.kraken()

portfolio = {
    'balance': 10000,
    'btc': 0,
    'trades': [],
    'starting_balance': 10000,
    'buy_price': 0  # Kahan pe buy kiya — track karo
}

STOP_LOSS_PCT = 5    # 5% neeche gira toh sell
TAKE_PROFIT_PCT = 10 # 10% upar gaya toh sell

def get_rsi():
    ohlcv = exchange.fetch_ohlcv('BTC/USD', '1h', limit=100)
    df = pd.DataFrame(ohlcv, columns=['time','open','high','low','close','volume'])
    df['rsi'] = ta.momentum.RSIIndicator(df['close'], window=14).rsi()
    price = df['close'].iloc[-1]
    rsi = round(df['rsi'].iloc[-1], 2)
    return price, rsi

def print_portfolio(price):
    if portfolio['btc'] > 0:
        current_value = portfolio['btc'] * price
        # Stop-loss aur take-profit levels dikhao
        sl_price = round(portfolio['buy_price'] * (1 - STOP_LOSS_PCT/100), 2)
        tp_price = round(portfolio['buy_price'] * (1 + TAKE_PROFIT_PCT/100), 2)
        print(f"  Stop-loss at  : ${sl_price} (-{STOP_LOSS_PCT}%)")
        print(f"  Take-profit at: ${tp_price} (+{TAKE_PROFIT_PCT}%)")
    else:
        current_value = portfolio['balance']

    profit = round(current_value - portfolio['starting_balance'], 2)
    profit_pct = round((profit / portfolio['starting_balance']) * 100, 2)
    print(f"  Portfolio value : ${round(current_value, 2)}")
    print(f"  Total P&L       : ${profit} ({profit_pct}%)")

def execute_sell(price, reason):
    portfolio['balance'] = portfolio['btc'] * price
    portfolio['trades'].append({
        'type': f'SELL ({reason})',
        'price': price,
        'time': datetime.now().strftime('%H:%M:%S')
    })
    portfolio['btc'] = 0
    portfolio['buy_price'] = 0
    print(f"  *** SELL executed! Reason: {reason} ***")
    print(f"  BTC becha at ${price}")

def run_bot():
    print("=" * 45)
    print("SMART PAPER TRADING BOT")
    print(f"Starting balance : ${portfolio['starting_balance']}")
    print(f"Stop-loss        : {STOP_LOSS_PCT}%")
    print(f"Take-profit      : {TAKE_PROFIT_PCT}%")
    print("=" * 45)

    while True:
        now = datetime.now().strftime('%H:%M:%S')
        price, rsi = get_rsi()

        print(f"\n[{now}] BTC: ${price} | RSI: {rsi}")

        # BUY logic — RSI oversold
        if rsi < 35 and portfolio['btc'] == 0 and portfolio['balance'] > 0:
            portfolio['btc'] = portfolio['balance'] / price
            portfolio['buy_price'] = price
            portfolio['balance'] = 0
            portfolio['trades'].append({
                'type': 'BUY',
                'price': price,
                'time': now
            })
            print(f"  *** BUY executed! ***")
            print(f"  {round(portfolio['btc'], 6)} BTC kharida at ${price}")

        elif portfolio['btc'] > 0:
            change_pct = ((price - portfolio['buy_price']) / portfolio['buy_price']) * 100

            # Stop-loss — 5% neeche gira
            if change_pct <= -STOP_LOSS_PCT:
                execute_sell(price, "STOP-LOSS")

            # Take-profit — 10% upar gaya
            elif change_pct >= TAKE_PROFIT_PCT:
                execute_sell(price, "TAKE-PROFIT")

            # RSI overbought — normal sell
            elif rsi > 65:
                execute_sell(price, "RSI SELL")

            else:
                pct = round(change_pct, 2)
                print(f"  HOLD | Change from buy: {pct}%")

        else:
            print(f"  HOLD | RSI signal ka wait kar rahe hain...")

        print_portfolio(price)
        print(f"  Total trades: {len(portfolio['trades'])}")
        time.sleep(60)

run_bot()
```
